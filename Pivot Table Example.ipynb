```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a sample dataset
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-02', '2023-01-03', '2023-01-03', '2023-01-03'],
    'Product': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
    'Sales': [10, 15, 12, 20, 25, 22, 15, 20, 18],
    'Region': ['East', 'West', 'North', 'East', 'West', 'North', 'East', 'West', 'North']
}

df = pd.DataFrame(data)
df
```

```python
# Create a bar plot of sales by date and product
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='Date', y='Sales', hue='Product')

# Customize the plot
plt.title("Sales by Date and Product")
plt.xlabel("Date")
plt.ylabel("Sales")

# Display the plot
plt.show()
```

```python
pivot_table = df.pivot_table(index='Date', columns='Product', values='Sales', aggfunc='sum')
pivot_table
```

```python
# Create a stacked bar plot of the pivot table
ax = pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6), rot=0)

# Customize the plot
plt.title("Sales by Product and Date")
plt.xlabel("Date")
plt.ylabel("Sales")

# Add data labels to the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.text(x + width / 2, y + height / 2, f"{height}", horizontalalignment='center', verticalalignment='center')

# Display the plot
plt.show()
```
